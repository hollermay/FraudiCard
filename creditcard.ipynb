## Let's import the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('creditcard.csv')
df
#extracted top 5 and last 5 rows of the dataset
```

```python
df.info()
```

```python
df.isnull().sum()
# we have no missing values in the dataset
```

```python
# distribution of legit transaction and fraud transaction
df['Class'].value_counts()
```

unbalanced data 
0 -> Normal transaction
1-> Fraud

```python
# separating the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()  
```

Building a sample dataset having similar distribution of normal transactions and fraud transactions

```python
legit_sample = legit.sample(n=492)
# this randomly selects 492 samples from the legit dataset
```

concatenating two dataframes


```python
new_df = pd.concat([legit_sample, fraud], axis=0)
new_df.head()
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X =     new_df.drop(columns='Class', axis=1)
Y =    new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

```python
model = LogisticRegression()
```

Training model with training data

```python
model.fit(X_train, Y_train)
```

Evaluating our model based on accuracy score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
X_train
```

```python
print('Accuracy score (training data) : ', training_data_accuracy)
```

If accuracy score is more than 75% (or 0.75 )then the model is good

Training on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score (test data) : ', test_data_accuracy)
```

```python
from sklearn.metrics import f1_score

# Calculate F1 score for test set
f1_test = f1_score(Y_test, X_test_prediction)
print("F1 score for test set:", f1_test)

# Calculate F1 score for train set
f1_train = f1_score(Y_train, X_train_prediction)
print("F1 score for train set:", f1_train)
```
